## Get rid of columns that have the most missing data: background, broadcast_time, broadcast_day, trailer_url, title_english


```python
anime_data = anime_data.drop(columns =['background', 'broadcast_time', 'broadcast_day', 'trailer_url', 'title_english'])
```

Columns synopsis, main_picture, url, title_japanese, members, favorites and title_synonyms, episode duration, updated_at,or approved have have no relevance in terms of how popular an anime is, so get rid of them. Start_date, real_start_date and start_year all give me the year which is ultimately what I want, so drop start_date, real_start_date. The same could be said for end_date and real_end_date. Let's keep real_end_date between them.

```python
anime_data = anime_data.drop(columns =['synopsis','main_picture', 'url', 'title_japanese', 'title_synonyms','episode_duration','total_duration','approved','updated_at', 'members', 'favorites', 'start_date', 'real_start_date', 'end_date'])
```

### Get rid of anime row that doesn't have a score, we need the score to determine popularity

```python
anime_data = anime_data[anime_data['score'].notna()]
```

### Let's also get rid of the music type, since those are just music videos. 


```python
anime_data = anime_data[anime_data['type']!= 'music']
anime_data.head()
```

### We should get rid of any anime that has not yet aired at the time this dataset was created, as there will be no relevant data for these shows

```python
anime_data = anime_data[anime_data['status']!= 'not_yet_aired']
```

```python
missing = pd.concat([anime_data.isnull().sum(), 100 * anime_data.isnull().mean()], axis=1)
missing.columns=['count', '%']
missing.sort_values(by=['count'], ascending = True)
```

```python
anime_data['demographics'].value_counts()
```

## Let's fill in some of our demographic data
for any anime rated r or above let's put the demographic as "Adult" for any missing values.
For anime pg-13, pg and g we'll categorize it as 'kids, Shounen', as the g and pg will be for kids and shounen is the demographic that appears most based on value_counts

```python
anime_data['demographics']= anime_data['demographics'].fillna('other')
```

### Let's also fill in the data that has more than 10% of missing data, (source studios producers themes licensors), with "Unknown" as the fillin

```python
anime_data[['source', 'studios', 'producers', 'themes', 'licensors']] = anime_data[['source', 'studios', 'producers', 'themes', 'licensors']].fillna('Unknown')
```
